**1)** Descubra os seguintes polinômios de Taylor e plote a
aproximação obtida junto com a função original:

**a)** $ \sqrt[7]{x}$ com P1(x) (linear)

```python
x = np.arange(0,2,0.01)
f = x ** (1/7)
p1 = x/7 + 6/7

plt.plot(x,f, label='$f = \sqrt[7]{x}$')
plt.plot(x,p1, label='P1= x/7 + 6/7')

plt.legend()
plt.grid()
plt.show()
```

**b)** sen(x) com P2(x) (quadrático)

```python
x = np.arange((-1)*np.pi,np.pi,0.1)
f = np.sin(x)
p2 = x

plt.plot(x,f, label='f = sen(x)')
plt.plot(x,p2, label='P2 = x')

plt.legend()
plt.grid()
plt.show()
```

**2)** Produza uma fórmula geral para o polinômio de Taylor
expandido em torno de a = 0 de grau n, além disso
tomando n = 20 gere os gráficos dos polinâmio obtidos.

**a)** $\frac{1}{1-x}$

$P_n(x) = 1 + x + x^2 + ... + x^n$

```python
def taylor(x, n):
    resposta = np.zeros(len(x))
    for i in range(n+1):
        resposta+=x**i
    return resposta

x= np.arange(-1,1,0.01)

plt.plot(x,taylor(x,20))

plt.grid()
plt.show()
```

**b)** $\sqrt{1+x}$ 

$P_n(x) = 1 + \frac{1}{2}x + \frac{\frac{1}{2}(\frac{1}{2}-1)x^2}{2!} + \frac{\frac{1}{2}(\frac{1}{2}-1)(\frac{1}{2}-2)x^3}{3!} + ... + \left(
\begin{array}{c}
1/2\\
n
\end{array}
\right)
x^n$

```python
def taylor(x, n):
    resposta = np.ones(len(x))
    derivada=0.5
    for i in range(1,n+1):
        resposta += derivada * (x ** i) / factorial(i)
        derivada *= (0.5 - i)
    return resposta

x = np.arange(-1,1,0.01)

plt.plot(x,taylor(x,20))

plt.grid()
plt.show()
```
