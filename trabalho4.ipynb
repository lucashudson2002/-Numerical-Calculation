1) Faça uma função que receba 2 vetores contendo os valores de x e os f(x) correspondentes e um valor a ser avaliado atraves de uma interpolação polinomial na forma de lagrange no Google Colab. 

Considere os seguintes valores e usando a função implementada faça:


```
x = np.array([55,70,85, 100, 115, 130]) #velocidade em km/h
```

a) Considerando uma interpolação de 2o grau aproxime 80km/h e 105km/h

b) Considerando uma interpolação de 3o grau aproxime 80km/h e 105km/h

c) Considerando todos os pontos aproxime 80km/h e 105km/h


````python
x = np.array([55,70,85, 100, 115, 130]) #velocidade em km/h
fx = np.array([14.08, 13.56, 13.28, 12.27, 11.3, 10.4]) #consumo em litro/km

def lagrange(n, x, y, z):
    r=0
    for i in range(n):
        c=1
        d=1
        for j in range(n):
            if i!=j:
                c=c*(z-x[j])
                d=d*(x[i]-x[j])
        r = r+y[i]*(c/d)
    return r
````

````python
grau2_80 = lagrange(3, x, fx, 80)
grau2_105 = lagrange(3, x, fx, 105)
print(f'Aproximando 80km/h no 2o grau, o consumo é de {grau2_80:.3f} litro/km')
print(f'Aproximando 105km/h no 2o grau, o consumo é de {grau2_105:.3f} litro/km')
````

````python
grau3_80 = lagrange(4, x, fx, 80)
grau3_105 = lagrange(4, x, fx, 105)
print(f'Aproximando 80km/h no 3o grau, o consumo é de {grau3_80:.3f} litro/km')
print(f'Aproximando 105km/h no 3o grau, o consumo é de {grau3_105:.3f} litro/km')
````

````python
todos_80 = lagrange(len(x), x, fx, 80)
todos_105 = lagrange(len(x), x, fx, 105)
print(f'Aproximando 80km/h com todos os pontos, o consumo é de {todos_80:.3f} litro/km')
print(f'Aproximando 105km/h com todos os pontos, o consumo é de {todos_105:.3f} litro/km')
````

````python
def x_t (a,b,t):
    return (b-a)*t/2 + (b+a)/2
````

````python
def dx_t(a,b):
    return (b-a)/2
````

````python
def F(a,b,t,f):
    return f(x_t(a,b,t))*dx_t(a,b)

def quad_gaus_2(f,a,b):
    F0 = F(a,b,-np.sqrt(3)/3,f)
    F1 = F(a,b,np.sqrt(3)/3,f)
    return F0+F1
````

````python
def quad_gaus_3(f,a,b):
    F0 = F(a,b,-np.sqrt(3/5),f)
    F1 = F(a,b,0,f)
    F2 = F(a,b,np.sqrt(3/5),f)
    return 5/9*F0+8/9*F1+5/9*F2
````
